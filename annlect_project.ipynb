# top 3 campaigns in terms of cpm( total media cost / total impressions *1000)

```python
media_cost_df = df_merge[df_merge['Metric'] == 'Media Cost'].copy()
media_cost_df.head(2)
media_cost_df.rename(columns={'Value': 'media_cost'},inplace = True)
media_cost_df
media_cost_df.drop(columns=['Metric'], inplace=True)
media_cost_df
```

```python
impressions_df = df_merge[df_merge['Metric'] == 'Impressions'].copy()
impressions_df.head(2)
impressions_df.rename(columns={'Value': 'impressions'},inplace = True)
impressions_df
impressions_df.drop(columns=['Metric'], inplace=True)
impressions_df
```

```python
cost_imp = pd.merge(media_cost_df, impressions_df, on = ['Date', 'Data source', 'Campaign', 'Media Type', 'City',
       'Month', 'Message type', 'Model'], how= 'inner')
cost_imp
df_media = cost_imp
```

```python
cost_imp= cost_imp.groupby(['Campaign'])[['media_cost', 'impressions']].sum().reset_index()
```

```python
cost_imp['cpm'] = (cost_imp['media_cost']/cost_imp['impressions'])*1000
#CPM.sort_values(ascending = False).head(3)
cost_imp.sort_values(by = 'cpm',).sort_values(by = 'cpm').head(3)
```
